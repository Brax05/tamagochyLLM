# MascotaLM — Chateá con Mochi 🐾

Descargá el modelo pre-entrenado de 9M parámetros y chateá con Mochi, tu mascota virtual.

**Modelo:** [Brax055/mascotalm-9M](https://huggingface.co/Brax055/mascotalm-9M)

Ejecutar todos los cells — no necesita GPU.

```python
# Setup + Descarga
!pip install -q torch tokenizers huggingface_hub
import os, shutil
if os.path.exists('/content/mascota'): shutil.rmtree('/content/mascota')
os.makedirs('/content/mascota'); os.chdir('/content/mascota')

from huggingface_hub import snapshot_download
snapshot_download(repo_id='Brax055/mascotalm-9M', local_dir='.')
print('Modelo descargado.')
```

```python
# Cargar modelo
from inference import MascotaInference
import torch

engine = MascotaInference('pytorch_model.bin', 'tokenizer.json',
                          device='cuda' if torch.cuda.is_available() else 'cpu')

def chat(prompt):
    return engine.chat_completion(
        [{'role': 'user', 'content': prompt}], max_tokens=64
    )['choices'][0]['message'].get('content', '').strip()

# Test rápido
for p in ['hola mochi', '¿tenés hambre?', 'contame un chiste',
          '¿qué son los impuestos?', 'buenas noches mochi']:
    print(f'Vos: {p}')
    print(f'Mochi: {chat(p)}')
    print()
```

```python
# Chat interactivo — escribí tus mensajes
while True:
    try:
        p = input('Vos> ').strip()
    except (KeyboardInterrupt, EOFError):
        break
    if not p or p.lower() in ('chau', 'salir', 'exit', 'q'):
        print('Mochi> bueno. chau. volvé pronto.'); break
    print(f'Mochi> {chat(p)}\n')
```
